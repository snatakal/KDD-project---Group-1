```python
import pandas as pd
import warnings
import os
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
import joblib
```

```python
parent_dir = os.path.dirname(os.getcwd())
out_path = os.path.join(parent_dir, 'data', 'transform')
```

```python
tf_df1 = pd.read_csv(os.path.join(out_path, 'tf_df1.csv'))
tf_df2 = pd.read_csv(os.path.join(out_path, 'tf_df2.csv'))
tf_df3 = pd.read_csv(os.path.join(out_path, 'tf_df3.csv'))
tf_df4 = pd.read_csv(os.path.join(out_path, 'tf_df4.csv'))
tf_df5 = pd.read_csv(os.path.join(out_path, 'tf_df5.csv'))
```
